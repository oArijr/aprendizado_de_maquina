3) Porque utilizamos a classe chamada Sequential para a RNA?

```python
rede_neural = Sequential()


rede_neural.add(Input(shape=(30,)))
rede_neural.add(Dense(
    16, # Uma convencao usada que dividimos a soma da entrada e saida por 2 (30 + 1 / 2 = 15,5 = 16)
    activation='relu',
    kernel_initializer=initializers.RandomUniform() # Inicializa os pesos dos neurônios com valores aleatórios
    )
)


rede_neural.add(Dense(1, activation='sigmoid'))
# Utilizamos sigmoid pois nossa saida varia entre 0 e 1 (Probabilidade de câncer de mama)
```

```python
# 3)
# R: Uma rede neural em sequencia linear, nossos dados de entrada têm uma única entrada
# e cada camada tem exatamente uma entrada e uma saída
```

4) A partir da RNA gerada, explique o que são os valores apresentados na tabela da rede_neural.summary()

```python
rede_neural.summary()

# A tabela exibida pelo summary() mostra a arquitetura da rede neural, incluindo:
# - As camadas utilizadas
# - O formato da saída gerada por cada camada
# - A quantidade de parâmetros treináveis (pesos e viéses)

# "Param #" indica o total de parâmetros treináveis em cada camada.
# Esse valor é a soma dos pesos e dos viéses (bias) de cada neurônio da camada.
# Por exemplo:
#   - dense_3: (30 entradas × 16 neurônios) + 16 viéses = 480 + 16 = 496
#   - dense_4: (16 entradas × 1 neurônio) + 1 viés = 16 + 1 = 17

# "Total params": Soma total dos parâmetros treináveis da rede: 496 + 17 = 513
# "Trainable params": Parâmetros que serão atualizados durante o treinamento (513)
# "Non-trainable params": Parâmetros fixos (não ajustáveis), nesse caso: 0
```

5) Adicione um otimizador Adam (https://www.tensorflow.org/api_docs/python/tf/keras/optimizers) e especifique a classe loss binário - binary crossentropy (https://www.tensorflow.org/api_docs/python/tf/keras/losses) e a classe metrics para utilizar a métrica de avaliação de acurácia binária (https://www.tensorflow.org/api_docs/python/tf/keras/metrics)

6) Para que servem os otimizadores? Como o otimizador Adam funciona?

7) Depois de estruturados os parâmtros da RNA, utilize a função .fit para fazer o treinamento da rede. Como foi utilizado o otimizador Adam, e ele é baseado na descida do gradiente estocástica, é possível definir a quantidade de registros que serão enviados para a RNA, isto é, em cada batch serão utilizados 10 registros.
```python
# 5)
rede_neural.compile(
    optimizer=tf.keras.optimizers.Adam(),  # Otimizador Adam
    loss=tf.keras.losses.BinaryCrossentropy(),  # Função de perda binária
    metrics=[tf.keras.metrics.BinaryAccuracy()]  # Métrica
)



# 6)
# Os otimizadores ajustam automaticamente os pesos da rede neural para que ela cometa cada vez menos erros ao fazer previsões.

# Adam ou Adaptive Moment Estimation. Combina Momentum e o RMSProp.


# Otimizacoes de descida de gradiente:

## Momentum:
# Momentum e o impulso. Se uma descida ocorre varias vezes, a 'velocidade' aumenta e ocorre a descida de forma mais rapida. Acumula a direção dos gradientes, melhorando performance.

# Imagine que você está empurrando uma bola por um terreno com declives. Se o terreno for sempre inclinado para o mesmo lado, a bola acelera.

## RMSProp:
# RMSProp ou Root Mean Square Propagation, ajusta a taxa de aprendizado ('tamanho do passo') de cada parâmetro individualmente.
# Tornar o passo mais estável em regiões com gradientes grandes e pequeno em regiões instáveis, x desce muito e y desce pouco.


## Adam ajusta a direção e o tamanho do passo



# https://www.youtube.com/watch?v=MD2fYip6QsQ



# 7) e 8)
rede_neural.fit(
    X_treinamento, y_treinamento,
    batch_size=10,  # Quantos exemplos de treinamento o modelo vai usar antes de atualizar os pesos, aka fazer backpropagation da media de 10 treinamento
    epochs=100 # Número de épocas de treinamento
)
# Serao utilizamos 43 batches (42 de 10 + 1 de 7)
```

13) Assista o vídeo https://youtu.be/RczbeFs_WbQ?si=f1Yu4LZaGDfZZpbZ para compreender melhor sobre como funciona o k-fold cross validation. Explique como esta rede foi configurada e como é possível chegar no resultado dela. O que é necessário fazer?


14) Calcule também o Desvio Padrão dos resultados para avaliar o modelo. O que é possível concluir com esse resultado?



```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

```python
def criar_rede():
  k.clear_session()
  rede_neural = Sequential([
      tf.keras.layers.InputLayer(shape=(30,)),
      tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
      tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
      tf.keras.layers.Dense(units=1, activation = 'sigmoid')])
  otimizador = tf.keras.optimizers.Adam(learning_rate = 0.001, clipvalue = 0.5)
  rede_neural.compile(optimizer = otimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
  return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs = 100, batch_size = 10)
# Utilizamos o KerasClassifier para depois poder realizar com a biblioteca scikeras a validação cruzada da rede.
```

```python
resultados = cross_val_score(estimator = rede_neural, X = X, y = y, cv = 10, scoring = 'accuracy')
# Divide 10 particoes, 90% treino e 10% teste
```

```python
print(f"Acurácias por fold: {resultados}")
print(f"Acurácia média: {resultados.mean() * 100:.2f}%")
print(f"Desvio padrão: {resultados.std() * 100:.2f}%")

# Nossa acuracia anteriormente sem cross validation havia sido de 77.19%, porem usando ele podemos ver que ela nao e tao ruim
# pois ficou com uma acuracia media no cross validation de 82.42%

# Se o modelo estivesse bem generalizado, a performance seria mais consistente entre os folds.
```

## Overfitting e Dropout

14) Como vimos, o modelo anterior sofreu overfitting. Para reduzir esse problema, aplicamos a técnica de regularização conhecida como dropout (artigo oficial: https://www.cs.toronto.edu/~rsalakhu/papers/srivastava14a.pdf e explicação em português: https://www.deeplearningbook.com.br/como-funciona-o-dropout/ ). Aplique o dropout de 20% na primeira e segunda camada oculta. O que acontece com os resultados? E o Desvio Padrão?

```python
def criar_rede_dropout():
    k.clear_session()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape=(30,)),
        Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
        tf.keras.layers.Dropout(0.2),  # Dropout após a primeira camada oculta
        Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
        tf.keras.layers.Dropout(0.2),  # Dropout após a segunda camada oculta
        Dense(units=1, activation='sigmoid')
    ])
    otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
    rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede_dropout, epochs = 100, batch_size = 10)
```

```python
resultados = cross_val_score(estimator = rede_neural, X = X, y = y, cv = 10, scoring = 'accuracy')
```

```python
print(f"Acurácias por fold: {resultados}")
print(f"Acurácia média: {resultados.mean() * 100:.2f}%")
print(f"Desvio padrão: {resultados.std() * 100:.2f}%")

# Dropa aleatoriamente 20% dos neurônios da camada anterior a cada batch.
```

## Tuning dos Hiperparâmetros (técnicas de ajuste)

Hyperparameter Tuning (Ajuste de Hiperparâmetros) é o processo geral de encontrar os melhores valores para os hiperparâmetros de um modelo de machine learning.
Exemplos de hiperparâmetros:

- Taxa de aprendizado (learning rate)

- Número de árvores em um RandomForest

- Número de camadas ou neurônios em uma rede neural

- Tipo de kernel em um SVM

O tuning pode ser feito por diferentes técnicas, como:

- Grid Search

- Random Search

- Bayesian Optimization, etc.

15) Descreva como a RNA foi configurada para fazer o processo de tuning.

```python
def criar_rede(optimizer, loss, kernel_initializer, activation, neurons):
  k.clear_session()
  rede_neural = Sequential([
      tf.keras.layers.InputLayer(shape=(30,)),
      tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
      tf.keras.layers.Dropout(rate = 0.2),
      tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
      tf.keras.layers.Dropout(rate = 0.2),
      tf.keras.layers.Dense(units=1, activation = 'sigmoid')])
  rede_neural.compile(optimizer = optimizer, loss = loss, metrics = ['binary_accuracy'])
  return rede_neural
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_treinamento_scaled = scaler.fit_transform(X_treinamento)
X_teste_scaled = scaler.transform(X_teste)

# Padronizamos os dados, transformando-os para que tenham média igual a 0 e desvio padrão igual a 1.
# Isso é conhecido como escala z ou padronização.
# O StandardScaler transforma os dados para que tenham média igual 0 e desvio padrão igual a 1.
# Fórmula aplicada z = (x - u) / s (conforme documentação https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)
print(X_treinamento_scaled)
```

```python
classificador_novo = tf.keras.models.load_model('/content/classificador_breast.keras')

# Treinar o modelo com os dados normalizados (sem early stopping)
history = classificador_novo.fit(
    X_treinamento_scaled,
    y_treinamento,
    validation_split=0.2,  # 20% para validação
    batch_size=10,  # Mantendo o parâmetro do GridSearch anterior
    epochs=50,  # Número fixo de épocas
    verbose=1
)
```

```python
from sklearn.metrics import classification_report

limiar = 0.5

y_pred = (classificador_novo.predict(X_teste_scaled) > limiar).astype(int)  # Previsões binárias
print("\nRelatório de Classificação no Conjunto de Teste:")
print(classification_report(y_teste, y_pred))

# O dataset tem 30 características numéricas, mas os valores apresentam escalas muito diferentes,
# Ex: radius_mean varia de 6.98 a 28.11, enquanto smoothness_mean está na ordem de 0.05 a 0.16).
# Além disso, o código mostra alguns valores discrepantes (como radius_mean de 9029, o que parece indicar vários outliers).

# Por isso fizemos a padronização. Então para testar traçamos um limiar de 0.5 (50%) de confiança para ser cancer de mama e alçamos os
# resultados abaixo.

# Pelo que entendemos 1 = maligno e 0 = benigno, a precisão maior para maligno de 0.97 (97%) é ótimo para diagnósticos médicos reais. Quando o modelo prevê um tumor como maligno, ele está correto em 97% dos casos.
# Além de ter o Recall bem elevado para ambas classes 0.96 (96%) e 0.95 (95%), que minimiza falsos negativos e é ideal para casos de diagnósticos médicos reais também.
```
